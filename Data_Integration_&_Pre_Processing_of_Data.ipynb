# What Is Data Integration?
Data integration combines various types and formats of data from any source across an organization into a data lake or data warehouse to provide a unified fact base for analytics. Working from this one data set allows businesses to make better decisions, aligns departments to work better together, and drives better customer experience.







# What is meant by data integration?
Data integration means consolidating data from multiple sources into a single dataset to be used for consistent business intelligence or analytics.

This is a very simple explanation for a complex topic that has evolved over its 30 year history. Understanding how data integration has transitioned from a backend, retrospective process into core real-time infrastructure starts with an examination of how data integration works.

# What are data integration tools?
Data Integration tools are software-based tools that ingest, consolidate, transform, and move data from source(s) to destination, performing mappings, transformations, and data cleansing along the way. Ultimately, they integrate the data into a single source of truth destination, such as a data lake or data warehouse. This allows consistent, reliable data for use in analytics and business intelligence.

# Pre-Processing of Data

**What Is Data Preprocessing?**

Data preprocessing is a step in the data mining and data analysis process that takes raw data and transforms it into a format that can be understood and analyzed by computers and machine learning.

Raw, real-world data in the form of text, images, video, etc., is messy. Not only may it contain errors and inconsistencies, but it is often incomplete, and doesnt have a regular, uniform design.

Machines like to process nice and tidy information  they read data as 1s and 0s. So calculating structured data, like whole numbers and percentages is easy. However, unstructured data, in the form of text and images must first be cleaned and formatted before analysis.

**Data Preprocessing Steps**

Let's checking for null values in our data

```python
tabel = pd.DataFrame({
    'Unique':df.nunique(),
    'Null':df.isnull().sum(),
    'NullPercent':df.isnull().sum() / len(df),
    'NaNN':df.isna().sum(),
    'Type':df.dtypes.values,
})
tabel
```

While merging the 2 dataframes an empty duplicate null column has been created. So, let's remove that column 1st which has no column name.

```python
null_column_index = None
for i, col in enumerate(df.columns):
    if df[col].isnull().all():
        null_column_index = i
        break

# Drop the null column from the DataFrame
df = df.drop(df.columns[null_column_index], axis=1)
```

# Imputation for missing values with Mean

As the missing values in our data set is continuous but not random, Mean imputation is the most realible. 

```python
#For features having less than 20% null values are being replaced with the mean value of that feature

columns_to_replace_with_mean = {}
for col in df.columns:
  null_exist = len(df[df[col].map(lambda x: x in ["m", "M", "NC"])])
  if null_exist:
    df_copy = df.copy()
    df_copy = df_copy.drop(df_copy[df_copy[col].map(lambda x: x in ["m", "M", "NC"])].index)
    df_copy[col] = df_copy[col].astype(float)
    mean_value = round(df_copy.loc[:, col].mean(), 2)
    df[col] = df[col].replace(['M', 'm', 'NC'],[mean_value, mean_value, mean_value])
    df[col] = df[col].astype(float)

df
```

**Now we need to create a dependent variable 'LES_snowfall'. The snowfall prediction is mainly dependent on clouds, Temperature & Precepitation when temp<32F and precip > 0.**

```python
df['LES_snowfall'] = 0

df.loc[(df['Temp (F)'] <= 32.0) & (df['Precip (in)'] > 0.01) & (df['LES_Cloud_Exist'] == 'Y'), "LES_snowfall"] = 1
```

# Feature reduction

By looking at above table, we're going to remove all the correlated features with correlation more than 80% also temp, precip, LES_Cloud_Exist column as output feature (LES_snowfall) is build based on these features.

```python
df = df.drop(columns= ['Temp (F)', 'Precip (in)', 'LES_Cloud_Exist'])
```

As correlation matrix is an inverted triangle, both the half's has same values. So, taking the upper half to measure the correlation by keeping the values in dataframe.

```python
import numpy as np 

cor_matrix = df.corr().abs()
print(cor_matrix)

upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))
print(upper_tri)
```

Getting the list of columns with more than 80% correlation

```python
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.80)]
print(to_drop)
```

Dropped all the highly correlated features as dimensionality reduction is one key aspect in SOM. So, we're reducing the data redundancy.

```python
df = df.drop(columns = to_drop, axis=1)
df.head()
```

# Scaling the data to remove bias



```python
from sklearn.preprocessing import MinMaxScaler

X = df.drop(columns='LES_snowfall')
y = df['LES_snowfall']

scaler = MinMaxScaler(feature_range=(0,1))
X = scaler.fit_transform(X)
```

**Balancing dependent variable**

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X, y = smote.fit_resample(X, y)
ros_chd_plot=y.value_counts().plot(kind='bar')
plt.show()
```
